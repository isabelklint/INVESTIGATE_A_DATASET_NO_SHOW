## Introduction

### Dataset Description 

The focus of this investigation is the characteristics missed appointments. This No-show dataset from Kaggle provides the details of 100k medical appointments in Brazil. Specifically, the data holds information about appointments: to which hospital they were scheduled, demographic and health data of the patients, and whether or not appointments were kept.

My analyses are all focussed on the dependent variable: missed appointment. I will look at missed appointments by length of time between scheduling and appointment, by demographic features like patient neighbourhood and age group, and by patient health factors and age.

Here are the column names and a brief description of each:

- PatientId: A unique ID for each patient</li>
- AppointmentID: A unique ID for each appointment</li>
- ScheduledDay: The day the patient made their appointment</li>
- AppointmentDay: The day of the appointment</li>
- Gender, Age: patient demographic information</li>
- Neighbourhood: The location of the hospital</li>
- Scholarship:  ‘Yes’ denotes that the patient is using Brasilian welfare (Bolsa Família)</li>
- Hipertension, Diabetes, Alcoholism, Handcap: health factors. Presumably ‘Handcap’ refers to disability status</li>
- SMS_received: Whether the patient received a text reminder. (does this correlate with scholarship? do poorer people not get text reminders?)</li>
- No-show: Whether the patient missed an appointment (Yes) or kept an appointment (No)</li>

```python
# import statements

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import dask.dataframe as dd
import seaborn as sb

# 'magic word'
%matplotlib inline
```

## Data Wrangling

The No-show data comes from Kaggle and is provided by Udacity. Here are the first few rows of data, and the total number of rows and columns.

```python
# data from Kaggle and Udacity: noshowappointments-kagglev2-may-2016.csv

# df = pd.read_csv('/Database_No_show_appointments/noshowappointments-kagglev2-may-2016.csv')
# df.head()
# error: dataset too large

# use dask
df_dd = dd.read_csv("Database_No_show_appointments/noshowappointments-kagglev2-may-2016.csv", assume_missing=True)
df_dd.head()
```

```python
# change the dataframe type to continue using pandas
df = df_dd.compute()
x = df.shape
print('This dataset comprises this many rows and columns:', x)
```

The focus of this analysis is missed appointments. The percentage of appointments that were missed:

```python
# how many missed appointments are there?
# no-show info change to numbers
df_missed_percentage = df
df_missed_percentage['No-show'] = df_missed_percentage['No-show'].astype('str')
df_missed_percentage['No-show'].replace({'No': 1, 'Yes': 0}, inplace=True)
```

```python
# get the number of missed appointments (no-show = no)
missed = sum(df_missed_percentage['No-show'] == 0)

# get the number of appointments
total = df_missed_percentage['No-show'].count()
```

```python
# get percentage of total appointments that were missed
percent_attendance = missed/total
# can this be rounded?
percent_attendance2 = percent_attendance.round(decimals = 2)
print('The percentage of total appointments that were missed:',
percent_attendance2)
```

Here is a overview of the data:

```python
# quick overview of data
df.info()
```

```python
# look for duplicated rows
d = sum(df.duplicated())
print('There are', d, 'duplicated rows.')
```

```python
# look for duplicates where none should be
s = sum(df['AppointmentID'].duplicated())
print('Appointment IDs are duplicated', s, 'times.')
```

```python
s = sum(df['PatientId'].duplicated())
print('Patient IDs are duplicated', s, 'times.')
```

Some patients have multiple appointments.

Some column names and datatypes need to be changed.

First, Scheduled day and Appointment Day are 'object' datatypes. I changed these to Date-time. In the analysis section, I examine some assumptions about time of day of the appointment and distance in time between scheduling and appointment date.


```python
# change day columns to datetime type
df[['ScheduledDay', 'AppointmentDay']] = df[['ScheduledDay', 'AppointmentDay']].apply(pd.to_datetime)
df['ScheduledDate'] = df['ScheduledDay'].dt.date
df['AppointmentDate'] = df['AppointmentDay'].dt.date
df_time = df
```

```python
# drop any appointments that were scheduled after the appointment date:
df = df_time.loc[df_time['ScheduledDate'] <= df_time['AppointmentDate']] 

# get the difference in time between scheduling and the appointment date
df_time['Time_Difference'] = (df.AppointmentDate - df.ScheduledDate).astype('timedelta64[D]')
df = df_time
df_time.head
```

Next, the No-show column and values are changed to be less confusing. In the original data, a 'no' in the 'No-show' column indicates a kept appointment and a 'Yes' indicates a missed appointment. A 'yes' in a row about appointments indicating that the appointment was kept is clearer.

The different capitalisation between PatientId and AppointmentID is confusing. Those column headers are changed. Also, the values for PatientId and AppointmentID are too long for easy reading.

Hypertension is spelled as 'hipertension': that has been changed.  As column headers, 'Disability' is preferrable to 'Handcap' and 'Assistance' is clearer than 'Scholarship'. 'Neighbourhood' refers to the location of the hospital, so that has been changed to 'Hospital'. Also, underscores were added where a space would naturally go to make the headers easier to read.

I changed the datatype of the values for 'Gender' to 1 and 0, instead of M and F.

```python
# change column headings
df = df.rename(columns = {'AppointmentID': 'Appointment_ID',
                          'PatientId': 'Patient_ID',
                          'AppointmentDay': 'Appointment_Day',
                          'ScheduledDay': 'Scheduled_Day',
                          'Scholarship': 'Assistance',
                          'Neighbourhood': 'Hospital',
                          'Hipertension': 'Hypertension',
                          'Handcap': 'Disability',
                          'No-show': 'Attendance'})

# actually attendance values were changed above
# df_missed_percentage['No-show'].replace({'No': 1, 'Yes': 0}, inplace=True)
```

```python
# should i make this categorical?
df['Gender'] = df['Gender'].astype('str')
df["Gender"].replace({"M": 1, "F": 0}, inplace=True)
df.dtypes
```

Patient IDs are needlessly in scientific notation. The patient ids need to be unique, but their actual value is probably irrelevant. As they were randomly assigned, I changed them to integers.

The values for Appointment IDs and for Neighbourhoods are changed to numerical, for ease of reading.

I removed the patient outliers by age. Here is a boxplot representing the age column, and a second with the outliers removed.

```python
# look for outliers on box plot.
df.boxplot('Age')
```

```python
# get the mean and the standard deviation using numpy
df_mean, df_std = np.mean(df), np.std(df)
# identify ages beyond the standard deviation
outside_the_box = df_std * 3
lowest, highest = df_mean - outside_the_box, df_mean + outside_the_box
# make a df that only has ages in the box
df.drop(df[df['Age'] > highest['Age']].index, inplace = True)
df.drop(df[df['Age'] < lowest['Age']].index, inplace = True)
```

```python
# checking if patient_ID correlates to age in any way
# this was taking too long so i made a sample of the data to check
df_sample = df.sample(n=500).head(10)

def getcorr(df, column_1, column_2):
    x = column_1
    y = column_2
    corr = x.corr(y)
    print('The correlation is', corr2)

getcorr(df_sample, df_sample['Patient_ID'], df_sample['Age'])
```

The patient ID correlates strongly to age, this indicates that they are assigned differently based on the age of the patient. I examine the difference between adults and children in the analysis section.

The float columns are  categories, so the data is changed to reflect that.

```python
# change to easier numbers
for column in ['Appointment_ID',
'Hospital',
'Patient_ID','Disability']:
    df[column] = df.groupby(column).ngroup() + 1

# first change from string to float
df['Attendance'] = df['Attendance'].astype(float, errors = 'raise')

# change to categorical
for column in ['Assistance',
               'Hypertension',
               'Diabetes',
               'Alcoholism',
               'SMS_received',
               #'Attendance',
               'Gender']:
    df[column] = df[column].astype('category')
x = df.shape
print(x)
```

The data is cleaned. 

Column headers were changed to be more clear. Datatypes were changed for easier reading and analysis. Also, columns with two values are now categorical, and some outliers were removed.

The data is now ready for analysis.

### Which appointments are being missed? Appointment Questions Analysis

Appointments might be missed for reasons related to when the appointment was scheduled or the day of the week of the appointment.

The longer the length of time between scheduling an appointent and the appointment date, the more likely it was that the appointment was missed.

The correlation between length of time between scheduling and keeping of an appointment:

```python
# what is correlation between difference and attendance?
getcorr(df, df['Time_Difference'], df['Attendance'])
```

The mean length of time between scheduling date and appointment date for missed appointments is approximately 16 days. For the appointments that were kept, the mean is about 8 days. If the time between scheduling and appointment date is about 8 days or less, appointments were more often kept than not. However, if the time between scheduling and appointment date was any longer, the appointments were more often missed.

The day of the week of the appointment affected whether or not someone attended their appointment. The correlation between day of the week and attendance is:

```python
df_days = df.reset_index()
df_days['Day_of_Week'] = df['Appointment_Day'].dt.dayofweek
getcorr(df_days, df_days['Day_of_Week'], df_days['Attendance'])
```

The correlation between the day of the week that the appointment falls on and whether or not a patient kept that appointment is weak.

### Whose appointments are being missed? Demographic questions analysis

Certain demographic groups might be more likely to miss appointments.

The mean age of those who keep or miss appointments seems about the same.

```python
# look at mean age by
# time distance and keeping appointments
df_attendance_age_time = pd.DataFrame(df,columns=['Attendance', 'Time_Difference', 'Age'])
df_attendance_age_time.groupby(['Attendance']).describe(include = 'all')
```

However, a closer look shows that those who keep appointments tended to be young or old, not whereas those who missed appointments are more likely to be in their 20s, 30s, and 40s.

```python
# a more granular look at time difference and appointment
df_attendance_age_time = pd.DataFrame(df,columns=['Attendance', 'Time_Difference', 'Age'])
df_attendance_age_time.groupby(['Attendance', 'Time_Difference']).describe(include = 'all')
```

At a glance, it appears that babies make their appointments more than anyone.

```python
df_attendance_age_time.groupby(['Age', 'Attendance']).size().unstack().plot.bar(colormap='cool')
```

Regardless of age, more people made their appointment if they scheduled it within the previous 8 days. Otherwise, they were more likely to miss their appointment.

The location of the hospital is correlated to appointments attendance.

```python
# look at the difference in stats around 
# hospital, time difference, and keeping appointments
df_hospitals = pd.DataFrame(df,columns=['Attendance','Age', 'Hospital', 'Patient_ID'])
df_hospitals = df.reset_index()

#df.groupby(['Hospital', 'Attendance']).describe(include = 'all')
df_hospitals.groupby(['Attendance', 'Hospital']).size().unstack().plot.bar(colormap='winter', legend = False)
```

Some hospitals have lower attendance than others. 

Some hospitals are in neighbourhoods with a high level of patients with assistance. What is the correlation between neighbourhood and assistance?

```python
getcorr(df_sample, df_sample['Assistance'], df_sample['Hospital'])
```

```python
getcorr(df_sample, df_sample['Attendance'], df_sample['Assistance'])
```

There is a weaker but still robust correlation between patients with assistance and missed appontments.
